```python
#create function for plotting the alphabet images

def plot_alphabet(alphabet_data):
    """Plots alphabetic character images.
    """
    fig = plt.figure(figsize=(8,8))
    for i in range(alphabet_data.shape[0]):
        plt.subplot(alphabet_data.shape[0]//6 + 1, 6, i + 1)
        plt.subplots_adjust(hspace = 0.9, wspace = 0.2)
        a = alphabet_data[i, :].reshape((7, 5))
        plt.imshow(a)
```

```python
import h5py      #import h5py package to load .mat file format in new version of matlab mat files
with h5py.File('/content/drive/My Drive/Colab Notebooks/SOM/digits.mat', 'r') as file:
    digits = list(file['data'])
digits_numpy_list = np.asarray(digits)
print(digits_numpy_list.size)
digits_numpy_list = digits_numpy_list.reshape(10, 1100, 256)
print(digits_numpy_list.size)
digits_numpy_list = digits_numpy_list.transpose()

#split data to train and test
train = digits_numpy_list[:, :800, :]
test = digits_numpy_list[:, 800:, :]

train = np.transpose(train, (1, 2, 0))
test = np.transpose(test, (1, 2, 0))

train = train.reshape(8000, 256)
test = test.reshape(3000, 256)

  
print('train_size: ', train.shape)
print('test_size: ', test.shape)

```

```python
#plotting the sample image in dataset

plt.imshow(train[3].reshape(16,16).T)
```

```python
from sklearn.preprocessing import scale       #import for scaling data

train = scale(train)  #scaling

#create SOM networt with 30 * 30 grid size
som = MiniSom(30, 30, 256, sigma=4,
              learning_rate=0.5, neighborhood_function='triangle')
#som.pca_weights_init(train)
print("Training...")
som.pca_weights_init(train)  #initialize weights to speed up the convergence by PCA algorithm
som.train_random(train, 5000, verbose=True)  # random training
print("\n...ready!")

```

```python
#create labels to test the SOM network

num = []
for i in range(0, 8000, 10):
  num.append(0)
  num.append(1)
  num.append(2)
  num.append(3)
  num.append(4)
  num.append(5)
  num.append(6)
  num.append(7)
  num.append(8)
  num.append(9)
  
  
print(num)
num = np.asarray(num)
```

```python
# plotting the train data position after fit on SOM network Grid

plt.figure(figsize=(8, 8))
wmap = {}
im = 0
for x, t in zip(train, num):  # scatterplot
    w = som.winner(x)
    wmap[w] = im
    plt. text(w[0]+.5,  w[1]+.5,  str(t),
              color=plt.cm.rainbow(t / 10.), fontdict={'weight': 'bold',  'size': 11})
    im = im + 1
plt.axis([0, som.get_weights().shape[0], 0,  som.get_weights().shape[1]])
plt.savefig('/content/drive/My Drive/Colab Notebooks/SOM/train_som_digits.png')
plt.show()
```

#### As you can see, data with identical labels is located in areas close to each other. The reason for the error and the lack of placing data related to a particular label in a domain can be called a small network. And we're likely to get better precision with increasing dimensions. However, there may be a small error with a very large network

```python
#plotting the test data on SOM network grid

plt.figure(figsize=(8, 8))
wmap = {}
im = 0
for x, t in zip(test, num):  # scatterplot
    w = som.winner(x)
    wmap[w] = im
    plt. text(w[0]+.5,  w[1]+.5,  str(t),
              color=plt.cm.rainbow(t / 10.), fontdict={'weight': 'bold',  'size': 11})
    im = im + 1
plt.axis([0, som.get_weights().shape[0], 0,  som.get_weights().shape[1]])
plt.savefig('/content/drive/My Drive/Colab Notebooks/SOM/test_som_digits.png')
plt.show()
```

```python

import scipy.io      #import scipy to load .mat file format
mat = scipy.io.loadmat('/content/drive/My Drive/Colab Notebooks/SOM/Cities.mat')


#split data of cities
Cities1 = mat['Cities1']
Cities2 = mat['Cities2']
Cities3 = mat['Cities3']
Cities4 = mat['Cities4']
```

```python
#Cities 1

N_points = len(Cities1)
print(N_points)
N_neurons = N_points*

#plotting cities1 map
x = Cities1[:, 0:1].flatten()
y = Cities1[:, 1:2].flatten()
plt.figure(figsize=(20,10))
plt.scatter(x, y)
```

```python
#create SOM network with 1 * (2*len(cities1)) grid size

som = MiniSom(1, N_neurons, 2, sigma=3, neighborhood_function='triangle')
points = np.array([x,y]).T
som.pca_weights_init(points)    #initialize weights with PCA algorithm
som.train_batch(points, 100, verbose= True)      #training process
#som.train_random(points, 100)
#plotting result
plt.figure(figsize=(20,10))
plt.scatter(x,y)
visit_order = np.argsort([som.winner(p)[1] for p in points])
plt.plot(points[visit_order][:,0], points[visit_order][:,1])
for i, p in enumerate(points[visit_order]):
    plt.text(p[0]+.01, p[1]+.01, str(i+1))
```

```python
print(visit_order)
```

```python
#calculate the len of path

distance = 0
position = visit_order[0]
for index in visit_order[1:]:
    step_distance = np.sqrt((int(x[index]) - int(x[visit_order[index-1]]))**2 + (int(y[index]) - int(y[visit_order[index - 1]]))**2 )
    distance += int(step_distance)
    
print(distance)
    
```
